```python
import pandas as pd
import numpy as np
from numpy import nan as NA
df = pd.read_csv('https://raw.githubusercontent.com/JuJu-Ren/AIM-5001/main/M11_Data.csv', header=[0])
df
```

## 1.1Combining and reshaping data in Pandas to tidy and transform/reshape
* We have many 'NaN' values in the 'Month' column as a result of the melt() operation

```python
df["Month"].fillna(method='ffill', inplace = True)
df.head()
```

- We are defining the new name "suppliers" for the supplier names and "Value" for all the data value in each colomn

```python
# Melting from wide to long format, first 2 columns are used to 
ids = ["Month","Category"]
# the remaining columns are treated as data values to be pivoted into long format
values = df.columns[2:]
df = pd.melt(frame=df, id_vars=ids, value_vars=values, var_name='Suppliers', value_name='Value')
df.head()
```

- split the purchased and consumed data into two columns

```python
new = df["Value"].str.split(':', expand = True)
df["Purchased"],df["Consumed"]= new[0],new[1]
df.drop(columns = ["Value"], inplace = True)
df
```

#### The data is now in 'tidy' format:

- Each row is now comprised of a single observation

```python
data = pd.read_csv('https://raw.githubusercontent.com/JuJu-Ren/AIM-5001/main/bridges.data.version1.csv', header=None)
data.columns =["ID","River","Location","Erected","Purpose","Length","Lanes",
               "Clear-G","T/D","Material","Span","Rel-l","Type"]
data
```

## 2.1
how many bridges of each ‘Purpose’/’Material’ grouping within the data set have been constructed over each of the rivers listed in the data set.

```python
grouped = data["Material"].groupby([data["River"],data['Purpose'],data["Material"]]).agg(['count'])
grouped.rename(columns ={'count':'How Many?'})
```

## 2.2
show the average length for each ‘Purpose’/’Material’ bridge grouping within the data set

```python
# cleaning up the ‘Length’ column 
data["Length"] = data["Length"].replace('?', np.nan)
data =  data.astype({'Length':float})
# creating the report

# using "mean" to represent "average length", group by purpose, mertial in the requested order
grouped = data["Length"].groupby([data['Purpose'],data["Material"]]).agg(['mean'])
# rename the column
grouped.rename(columns ={'mean':'Average Length'})
```

## 2.3 
report that shows the average length, count, minimum length, and maximum length of bridges built during 4 equal length time periods (1818 – 1860; 1860-
1902; 1902-1944; 1944-1986).

```python
# check the "Erected"category, divide it with 4 equal length, set it as period
period = pd.cut(data['Erected'],4)
period[::30]
```

```python
def get_stats(group):
    return {'Average Length': group.mean(),'Count': group.count(), 'Max Length': group.max(),
            'Min Length': group.min()}
# short "Erected" with 4 cuts
grouped = data["Length"].groupby(period)
# change the names in multiple levels column 
grouped.apply(get_stats).unstack()
```
